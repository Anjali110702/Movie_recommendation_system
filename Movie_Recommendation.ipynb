```python
df_features = df[['genres','content_rating','movie_title','plot_keywords','director_name']].fillna(' ')
```

```python
x = df_features['genres']+ ' '+ df_features['content_rating']+ ' '+ df_features['movie_title']+ ' '+ df_features['plot_keywords']+ ' '+ df_features['director_name']+ ' ' 
```

```python
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(x)
```

```python
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score= cosine_similarity(X)
```

```python
Similarity_score
```
